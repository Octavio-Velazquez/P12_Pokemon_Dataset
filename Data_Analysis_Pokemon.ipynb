### 2. Leer el documento con los datos.

```python
# data.set_index("#") Tambien de este modo puedo indicar que una columna sea mi index.
# data.reset_index(inplace=True) Resetea el indice previamente establecido.
# data.sort_index() Ordena los valores del index de menor a mayor.
```

```python
def read_file(filepath):
    """Esta función lee el archivo indicado."""
    data = pd.read_csv(filepath, index_col="#")
    pd.set_option("display.max_columns", 12)
    return data
```

```python
ruta_documento = r"pokemon.csv"
datos = read_file(ruta_documento)
datos
```

```python
def file_info(readfile):
    """Check the basic info."""
    info_1 = readfile.head()
    print(f"\n1) - MUESTRA DE LOS PRIMEROS POKÉMON QUE APARECEN EN EL LISTADO:\n{info_1}")
    info_2 = readfile.tail()
    print(f"\n2) - MUESTRA DE LOS ÚLTIMOS POKÉMON QUE APARECEN EN EL LISTADO:\n{info_2}\n\n3) - INFORMACIÓN:")
    print(readfile.info())
    print(f"\n4) - COLUMNAS: \n{readfile.columns}")
    print(f"\n5) - FILAS Y COLUMNAS: \n{readfile.shape}")
    print(f"\n6) - TIPO DE DATOS: \n{readfile.dtypes}")
    print(f"\n7) - INDEX: \n{readfile.index}")
```

### 5. Filtrando los datos.

```python
datos
```

```python
# Aplico un filtro primero y luego busco usando loc, para seleccionar la columna name y generation.
filtro_nombre = datos["Name"] ==  "Pikachu"
datos.loc[filtro_nombre, ["Name", "Generation"]]
```

```python
# Uso el filtro para conocer los Pokemon, cuyo poder de ataque es mayor a 100.
filtro_ataque = datos["Attack"] > 100
datos.loc[filtro_ataque]
```

```python
# Aplicando dos filtros, para encontrar los Pokemon que sean del tipo Fire y Flying.
filtros_tipos = (datos["Type 1"] == "Fire") & (datos["Type 2"] == "Flying")
datos.loc[filtros_tipos]
```

```python
# Aplicando dos filtros, en donde se busca una opción u otra condición.
filtro_de_sp = (datos["Defense"] < 10) | (datos["Speed"] > 150)
datos.loc[filtro_de_sp]
```

```python
# Si pongo el simbolo ~ al lado del filtro,se busca lo contrario a los indicado en el filtro.
datos.loc[~filtro_de_sp]
```

```python
# Buscamos al Pokemon más poderoso. Aquel que tenga mayor cantidad en la columna Total.
filtro_power = (datos["Total"] > 700)
datos.loc[filtro_power]
```

```python
# Si quiero especificar las columnas que quiero visualizar.
datos.loc[filtro_power, ["Name", "HP", "Total", "Attack"]]
```

```python
# Usar una lista para filtrar datos especificos de una columna.

my_pokemon = ['Bulbasaur', 'Ivysaur', 'Venusaur', 'Charmander', 'Charmeleon', 'Charizard','Squirtle', 'Wartortle', 
              'Blastoise', 'Mega Blastoise', 'Caterpie', 'Metapod', 'Butterfree', 'Weedle', 'Kakuna', 'Beedrill',
              'Pidgey', 'Pidgeotto', 'Pidgeot', 'Rattata', 'Raticate', 'Spearow', 'Fearow', 'Ekans', 'Arbok',
              'Pikachu', 'Raichu', 'Sandshrew', 'Sandslash']


filtro_pkm = datos["Name"].isin(my_pokemon)
datos.loc[filtro_pkm]
```

```python
# Filtrando datos que son string o cadenas de texto.
filtro_tipo = datos["Type 2"].str.contains("Electric", na=False)
datos.loc[filtro_tipo]
```

#### Aplicando el método Apply()

##### El método Apply() solo actua sobre una Series. Es decir sobre una columna con multiples filas.

```python
datos["Digitos"] = datos["Pokemon"].apply(len)
```

```python
datos
```

```python
# Para poder usar métodos strings [lower, upper, etc.] con Apply(), tenemos que ponerlos en una función.
def update_type(type):
    """Convierte el texto en mayusculas."""
    return type.upper()
```

```python
datos["Type 1"].apply(update_type)
```

#### Rellenar por otros datos los valores NaN y luego usamos el método Apply()

```python
datos["Type 2"].fillna("-", inplace=True)
```

```python
def update_type2(type):
    """Comprueba."""
    if type == "-":
        return "No tiene un segundo tipo."
    else:
        return "Tiene dos tipos."
```

```python
datos["Type 2"].apply(update_type2)
```

```python
# Cuenta la cantidad de Pokemon que existen por el número de HP mostrado.
datos["HP"].value_counts()
```

```python
# Agrupamos por la columna HP. Y Buscamos un valor.
hp_grp = datos.groupby(["HP"])
hp_grp.get_group(100)
```

```python
# Agrupa a los Pokemon por tipo.
type_grp = datos.groupby(["Type 1"])
type_grp.get_group("Fire")
```

```python
# Muestra cada segundo tipo por el tipo principal.
type_grp["Type 2"].value_counts().head(55)
```

```python
# Muestra la cantidad de seguno tipo, que tienen los Pokemon de tipo de agua.
type_grp["Type 2"].value_counts().loc["Water"]
```

```python
# Muestra la mediana por ataque de cada tipo de Pokemon.
type_grp["ATTACK"].median()
```

```python
type_grp["ATTACK"].median().loc["Rock"]
```

```python
# Muestra la media y el promedio de los ataques por el tipo 1 de los Pokemon.
type_grp["ATTACK"].agg(["median", "mean"])
```
